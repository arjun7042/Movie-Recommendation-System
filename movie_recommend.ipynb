Now for cast, we will take only top 4 actor name (not the character name) and create tags on them for a particular movie.
So our function for extracting actor name will be a bit different

```python
def convert_cast(text):
    list_ = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 4:
            list_.append(i['name'])
            counter+=1
        else:
            break
    return list_
```

Now we want to extract only director name for a particular movie and we dont care about rest of the crew as we are including only director name in our tags.
So we will create a function to extract job=director and ignore other dictionaries

```python
def fetch_director(text):
    list_ = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            list_.append(i['name'])
    return list_ 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

Now we have to apply a transformation on genres, keywords, cast and crew to ensure there are no whitespaces between the word.
This is important for our system as we want them in tags as unique entity.
For example Sam can be considered as single entity and Worthington as another entity which are pointing towards same person, moreover we have another sam in crew column, so when we will combine all columns to create single tag column for a particular movie, our model will get confuse. Hence to avoid such situations, we are removing whitespaces to consider them as single entity.

```python
def remove_whitespace(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(remove_whitespace)
movies['crew'] = movies['crew'].apply(remove_whitespace)
movies['genres'] = movies['genres'].apply(remove_whitespace)
movies['keywords'] = movies['keywords'].apply(remove_whitespace)
movies.head()
```

Now we will combine overview, genres, keywords, cast and crew to create our single tag column

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

Now we have to define similarity based on these tags that we have created. So now we will vectorize this tag text to measure the similarity between tags of different movies. We will be using Cosine Similarity to measure the distance, but before that we will be vectorizing our text using BOW

### Text Vectorization using BOW

In this procedure, we will generate a large corpus which will contain all the words from all the tags from our dataset, then we will extract 6000 most common words (most occuring) and based on that we will create another dataframe containing the vectors for a particular movie.
note - We will be removing stopwords from our tags

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=6000,stop_words='english')
```

```python
vectors = cv.fit_transform(new['tags']).toarray()
```

## Cosine Similarity

Now we have 6000 dimensional vectors for each movie. So there are 4806 movies in total and we have 6000 dimensional vector for each movie. Now we will calculate cosine similarity(ranging 0 to 1) of one movie with all other movies based on these vectors.

+ larger cosine distance  -----> less similarity between movies ------> cosine similarity close to 0
+ smaller cosine distance -----> more similarity between movies ------> cosine similarity close to 1

Note- We are not using Euclidean distance as we have high dimensional data, so it will not be a reliable measure

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity.shape
```

```python
print(similarity[0]) #first movie's cosine similarity to all other 4806 movies
print(similarity[0].shape)
```

diagonal values in matrix will always be 1, as its cosine similarity of one movie with itself

```python
#Now we will recommend 5 movies closest to the movie we select
def recommend(movie):
    movie_index = new[new['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]

    for i in movies_list:
        print(new.iloc[i[0]].title)
        print(i[0])
```

```python
recommend('Batman Begins')
```
